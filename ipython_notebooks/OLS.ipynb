```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv('http://vincentarelbundock.github.io/Rdatasets/csv/datasets/longley.csv', index_col=0)
df.head()
df.plot(x='GNP', y='Employed', figsize=(12,6), kind='scatter')
pred = [51.8436 + 0.0348 * t for t in df['GNP'].values]
plt.plot(df['GNP'].values, pred, c='Red')
```

```python
X = sm.add_constant(df['GNP'])
est = sm.OLS(df['Employed'], X).fit()
print est.summary()
```
